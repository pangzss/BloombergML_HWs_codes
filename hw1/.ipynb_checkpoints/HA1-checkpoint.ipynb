##### Feature Normalization

```python
def feature_normalization(train, test):
    """Rescale the data so that each feature in the training set is in
    the interval [0,1], and apply the same transformations to the test
    set, using the statistics computed on the training set.

    Args:
        train - training set, a 2D numpy array of size (num_instances, num_features)
        test - test set, a 2D numpy array of size (num_instances, num_features)

    Returns:
        train_normalized - training set after normalization
        test_normalized - test set after normalization
    """
    # check constant features
    H,W = train.shape
    # each column respresents a feature, subtract the corresponding mean value
    # from each column to find the resulting all-zero column, which indicates
    # a feature that keeps at a constant value
    temp = train - np.expand_dims(np.mean(train,0),0)
    # find the non-zero columns
    idx = (temp != np.zeros((H,1)))
    # print(idx[0,:])
    # extract non-zero columns
    train_good = train[:,idx[0,:]]
    test_good = test[:,idx[0,:]]
    # Normalize to [0,1]
    feat_mins = np.amin(train_good,0,keepdims=True)
    feat_maxs = np.amax(train_good,0,keepdims=True)
    
    train_normalized = (train_good - feat_mins)/(feat_maxs-feat_mins)
    test_normalized = (test_good - feat_mins)/(feat_maxs-feat_mins)
    return train_normalized, test_normalized
```

##### The square loss function

```python
def compute_square_loss(X, y, theta,aug_mag=1):
    """
    Given a set of X, y, theta, compute the average square loss for predicting y with X*theta.

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D array of size (num_features+1), with the bias term
        aug_mag - the magnitude of the augmented dimension for the bias term, default 1.
    Returns:
        loss - the average square loss, scalar
    """
    if len(X.shape) == 1:
        X = np.expand_dims(X,0)
        y = np.expand_dims(y,0)
    loss = 0 #Initialize the average square loss
    X_aug = np.hstack( (X,aug_mag*np.ones((X.shape[0],1))) )
    loss = 1/X.shape[0]*(theta@X_aug.T@X_aug@theta-2*theta@X_aug.T@y+y@y)
    #TODO
    return loss
```

##### The gradient of the square loss function

```python
def compute_square_loss_gradient(X, y, theta,aug_mag=1):
    """
    Compute the gradient of the average square loss (as defined in compute_square_loss), at the point theta.

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D numpy array of size (num_features)

    Returns:
        grad - gradient vector, 1D numpy array of size (num_features)
    """
    if len(X.shape) == 1:
        X = np.expand_dims(X,0)
        y = np.expand_dims(y,0)
    X_aug = np.hstack( (X,aug_mag*np.ones((X.shape[0],1))) )
    grad = 1/X.shape[0]*(2*X_aug.T@X_aug@theta-2*X_aug.T@y)

    return grad
```

##### Gradient checker
Getting the gradient calculation correct is often the trickiest part of any gradient-based optimization algorithm. Fortunately, it's very easy to check that the gradient calculation is correct using the ndefinition of gradient.

```python
def grad_checker(X, y, theta, epsilon=0.01, tolerance=1e-4):
    """Implement Gradient Checker
    Check that the function compute_square_loss_gradient returns the
    correct gradient for the given X, y, and theta.

    Let d be the number of features. Here we numerically estimate the
    gradient by approximating the directional derivative in each of
    the d coordinate directions:
    (e_1 = (1,0,0,...,0), e_2 = (0,1,0,...,0), ..., e_d = (0,...,0,1))

    The approximation for the directional derivative of J at the point
    theta in the direction e_i is given by:
    ( J(theta + epsilon * e_i) - J(theta - epsilon * e_i) ) / (2*epsilon).

    We then look at the Euclidean distance between the gradient
    computed using this approximation and the gradient computed by
    compute_square_loss_gradient(X, y, theta).  If the Euclidean
    distance exceeds tolerance, we say the gradient is incorrect.

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D numpy array of size (num_features)
        epsilon - the epsilon used in approximation
        tolerance - the tolerance error

    Return:
        A boolean value indicating whether the gradient is correct or not
    """
    true_gradient = compute_square_loss_gradient(X, y, theta) #The true gradient
    num_features = theta.shape[0]
    approx_grad = np.zeros(num_features) #Initialize the gradient we approximate
    J = partial(compute_square_loss,X,y)
    for i in range(num_features):
        e_i = np.zeros(num_features)
        e_i[i] = 1
        approx_grad[i] = (J(theta + epsilon * e_i) - J(theta - epsilon * e_i) ) / (2*epsilon)
        
    temp = np.linalg.norm(approx_grad-true_gradient) < tolerance
    #print( np.linalg.norm(approx_grad-true_gradient))
    return temp
```

```python
### Generic gradient checker
def generic_gradient_checker(X, y, theta, objective_func, gradient_func, epsilon=0.01, tolerance=1e-4):
    """
    The functions takes objective_func and gradient_func as parameters. 
    And check whether gradient_func(X, y, theta) returned the true 
    gradient for objective_func(X, y, theta).
    Eg: In LSR, the objective_func = compute_square_loss, and gradient_func = compute_square_loss_gradient
    """
    true_gradient = gradient_func(X, y, theta) #The true gradient
    num_features = theta.shape[0]
    approx_grad = np.zeros(num_features) #Initialize the gradient we approximate
    J = partial(objective_func,X,y)
    for i in range(num_features):
        e_i = np.zeros(num_features)
        e_i[i] = 1
        approx_grad[i] = (J(theta + epsilon * e_i) - J(theta - epsilon * e_i) ) / (2*epsilon)
        
    temp = np.linalg.norm(approx_grad-true_gradient) < tolerance

    return temp
```

##### Batch Gradient Descent

```python
def batch_grad_descent(X, y, alpha=0.1, num_step=1000, grad_check=False):
    """
    In this question you will implement batch gradient descent to
    minimize the average square loss objective.

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        alpha - step size in gradient descent
        num_step - number of steps to run
        grad_check - a boolean value indicating whether checking the gradient when updating

    Returns:
        theta_hist - the history of parameter vector, 2D numpy array of size (num_step+1, num_features)
                     for instance, theta in step 0 should be theta_hist[0], theta in step (num_step) is theta_hist[-1]
        loss_hist - the history of average square loss on the data, 1D numpy array, (num_step+1)
    """
    num_instances, num_features = X.shape[0], X.shape[1]+1
    theta_hist = np.zeros((num_step+1, num_features)) #Initialize theta_hist
    print(theta_hist.shape)
    loss_hist = np.zeros(num_step+1) #Initialize loss_hist
    theta = np.zeros(num_features) #Initialize theta
    loss_hist[0] = compute_square_loss(X, y, theta)
    for i in range(num_step):
        grad = compute_square_loss_gradient(X, y, theta)
       # print(grad)
        if grad_check is True:
            assert grad_checker(X, y, theta, epsilon=0.01, tolerance=1e-4)
        theta = theta - alpha*grad 
        theta_hist[i+1] = theta
        loss_hist[i+1] = compute_square_loss(X, y, theta)
   
    return theta_hist,loss_hist
    #TODO
```

##### Backtracking Line Search

```python
def line_search(X, y, theta, objective_func, gradient_func, alpha=0.1,reg=False,lambda_reg=10**-2,aug_mag=1):
    if reg:
        J = partial(objective_func,X,y,
                    aug_mag=aug_mag)
        grad = gradient_func(X,y,theta,
                             lambda_reg=lambda_reg,aug_mag=aug_mag)
    else:
        J = partial(objective_func,X,y,
                   aug_mag=aug_mag)
        grad = gradient_func(X,y,theta,
                            aug_mag=aug_mag)    
    descent_dir = - grad/np.linalg.norm(grad)  # steepest descent direction
    m = grad@descent_dir
    
    c = 0.8
    tau = 0.8
    t = -c*m
    
    k=0
    while (J(theta) - J(theta+alpha*descent_dir) < alpha*t) and (k<100) :
        alpha *= tau
        k += 1
    
    return alpha
```

```python
def batch_plus_search(X, y, objective_func, gradient_func, alpha=0.1, num_step=1000, grad_check=False):
    """
    In this question you will implement batch gradient descent to
    minimize the average square loss objective.

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        alpha - step size in gradient descent
        num_step - number of steps to run
        grad_check - a boolean value indicating whether checking the gradient when updating

    Returns:
        theta_hist - the history of parameter vector, 2D numpy array of size (num_step+1, num_features)
                     for instance, theta in step 0 should be theta_hist[0], theta in step (num_step) is theta_hist[-1]
        loss_hist - the history of average square loss on the data, 1D numpy array, (num_step+1)
    """
    num_instances, num_features = X.shape[0], X.shape[1]+1
    theta_hist = np.zeros((num_step+1, num_features)) #Initialize theta_hist
    loss_hist = np.zeros(num_step+1) #Initialize loss_hist
    theta = np.zeros(num_features) #Initialize theta
    loss_hist[0] = compute_square_loss(X, y, theta)
    for i in range(num_step):
        grad = compute_square_loss_gradient(X, y, theta)
        if grad_check is True:
            assert grad_checker(X, y, theta, epsilon=0.01, tolerance=1e-4)
        alpha = line_search(X, y, theta, objective_func, gradient_func, alpha=alpha)
        theta = theta - alpha*grad 
        theta_hist[i+1] = theta
        loss_hist[i+1] = compute_square_loss(X, y, theta)
   
    return theta_hist,loss_hist
    #TODO
```

##### Regularization

```python
def compute_regularized_square_loss(X, y, theta,lambda_reg,aug_mag=1):
    """
    Given a set of X, y, theta, compute the average square loss for predicting y with X*theta.

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D array of size (num_features)

    Returns:
        loss - the average square loss, scalar
    """
    if len(X.shape) == 1:
        X = np.expand_dims(X,0)
        y = np.expand_dims(y,0)
    loss = 0 #Initialize the average square loss
    X_aug = np.hstack( (X,aug_mag*np.ones((X.shape[0],1))) )
    loss = 1/X.shape[0]*(theta@X_aug.T@X_aug@theta-2*theta@X_aug.T@y+y@y+lambda_reg*theta@theta)
    #TODO
    return loss
def compute_regularized_square_loss_no_bias(X, y, theta,lambda_reg,aug_mag=1):
    """
    Given a set of X, y, theta, compute the average square loss for predicting y with X*theta.

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D array of size (num_features)

    Returns:
        loss - the average square loss, scalar
    """
    if len(X.shape) == 1:
        X = np.expand_dims(X,0)
        y = np.expand_dims(y,0)
    loss = 0 #Initialize the average square loss
    X_aug = np.hstack( (X,aug_mag*np.ones((X.shape[0],1))) )
    loss = 1/X.shape[0]*(theta@X_aug.T@X_aug@theta-2*theta@X_aug.T@y+y@y+lambda_reg*theta[:-1]@theta[:-1])
    #TODO
    return loss
```

```python
def compute_regularized_square_loss_gradient(X, y, theta, lambda_reg,aug_mag=1):
    """
    Compute the gradient of L2-regularized average square loss function given X, y and theta

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D numpy array of size (num_features)
        lambda_reg - the regularization coefficient

    Returns:
        grad - gradient vector, 1D numpy array of size (num_features)
    """
    #TODO
    grad = compute_square_loss_gradient(X, y, theta,aug_mag=aug_mag)+ 2*lambda_reg*theta

    return grad
def compute_regularized_square_loss_gradient_no_bias(X, y, theta, lambda_reg,aug_mag=1):
    """
    Compute the gradient of L2-regularized average square loss function given X, y and theta

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D numpy array of size (num_features)
        lambda_reg - the regularization coefficient

    Returns:
        grad - gradient vector, 1D numpy array of size (num_features)
    """
    #TODO
    grad = compute_square_loss_gradient(X, y, theta,aug_mag=aug_mag)+ 2*lambda_reg*np.append(theta[:-1],0)

    return grad
```

```python
def regularized_grad_descent(X, y, objective_func, gradient_func,
                             alpha=0.05, lambda_reg=10**-2,aug_mag=1, num_step=1000,
                             search=False,obj_with_reg=False):
    """
    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        alpha - step size in gradient descent
        lambda_reg - the regularization coefficient
        num_step - number of steps to run
    
    Returns:
        theta_hist - the history of parameter vector, 2D numpy array of size (num_step+1, num_features)
                     for instance, theta in step 0 should be theta_hist[0], theta in step (num_step+1) is theta_hist[-1]
        loss hist - the history of average square loss function without the regularization term, 1D numpy array.
    """
    num_instances, num_features = X.shape[0], X.shape[1]+1
    theta = np.zeros(num_features) #Initialize theta
    theta_hist = np.zeros((num_step+1, num_features)) #Initialize theta_hist
    loss_hist = np.zeros(num_step+1) #Initialize loss_hist
    loss_hist[0] = objective_func(X, y, theta,
                                 aug_mag=aug_mag)
    for i in range(num_step):
        grad = gradient_func(X, y, theta, lambda_reg,
                             aug_mag=aug_mag)
        if search == True:
            alpha = line_search(X, y, theta, objective_func, gradient_func, 
                                alpha=alpha,reg=True,lambda_reg=lambda_reg,aug_mag=aug_mag)
        theta = theta - alpha*grad 
        theta_hist[i+1] = theta
        if obj_with_reg:
            loss_hist[i+1] = objective_func(X, y, theta,
                                       aug_mag=aug_mag,lambda_reg=lambda_reg)
        else:
            loss_hist[i+1] = objective_func(X, y, theta,
                                       aug_mag=aug_mag)
   
    return theta_hist,loss_hist
    #TODO
```

```python
losses_test = []
aug_mags=np.arange(1,10,2)
theta_mag = []
fig,axes = plt.subplots(nrows=3)
for aug_mag in aug_mags:
    theta_hist,loss_hist = regularized_grad_descent(X_train, y_train,
                             compute_square_loss,
                             compute_regularized_square_loss_gradient,
                             alpha=0.01, lambda_reg=0.01, aug_mag=aug_mag,num_step=5000)
    losses_test.append(compute_square_loss(X_test, y_test,theta_hist[-1],aug_mag=aug_mag))
    
    #biases = [aug_mag*np.abs(theta[-1]) for theta in theta_hist[1:]] / np.linalg.norm(theta_hist[1:],1, axis = 1) 
    biases = [aug_mag*np.abs(theta[-1]) for theta in theta_hist[1:]] 
    theta_mag = np.log(np.linalg.norm(theta_hist[1:],2, axis = 1))
    

    
    axes[0].plot(biases[500:],label='aug_mag= '+str(aug_mag))
    axes[0].set_xlabel('steps',fontsize='x-small')
    axes[0].set_ylabel('abs(bias)',fontsize='x-small')
    axes[1].plot(theta_mag,label='aug_mag= '+str(aug_mag))
    axes[1].set_xlabel('steps',fontsize='x-small')
    axes[1].set_ylabel('theta log(1-norm)',fontsize='x-small')
    
# calculate test loss when the bias term is seperated from the theta vector, for comparison.
theta_hist,loss_hist = regularized_grad_descent(X_train, y_train,
                             compute_square_loss,
                             compute_regularized_square_loss_gradient_no_bias,
                             alpha=0.01, lambda_reg=0.01, aug_mag=1,num_step=5000)    
loss_test_no_bias = compute_square_loss(X_test, y_test,theta_hist[-1],aug_mag=1)

biases = [np.abs(theta[-1]) for theta in theta_hist[1:]] 
axes[0].plot(biases[500:],label='no bias ',color='black')

axes[0].legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize='x-small')
axes[1].legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize='x-small')
axes[2].plot(aug_mags,losses_test,label='bias not seperated')
axes[2].plot(aug_mags,[loss_test_no_bias for elem in aug_mags],label='bias seperated')
axes[2].set_xlabel('aug_mag',fontsize='x-small')
axes[2].set_ylabel('test squared loss',fontsize='x-small')
axes[2].legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize='x-small')
plt.subplots_adjust(hspace=1)


```

From the first figure, we can observe that when B (aug\_mag) is larger than one, the bias term will asymptotically converge to the value as when the bias term is not regularized by seperating it from the $\mathbf{\theta}$ vector. While when B is exactly one, the bias term is obviously suppressed by the regularization.


From the second figure, we know that the change of B does not affect the norm of $\mathbf{\theta}$ under the same $\mathbf{\lambda}$. This depicts that depending how much the bias term is regularized, the other parameters will be adjusted automatically in order to satisfy the regularization constraint.


From the third figure, we know that when the bias term is regularzed, the test performance is actually the best. Maybe regularizing bias can help avoid overfitting, but it does not do any good when it comes to adapting to the center shift of $\mathbf{y}$. In other words, this maybe due to the fact that the number of test data is too low to reflect the true performance????

##### Stochastic gradient descent

```python
from sklearn.utils import shuffle
```

```python
def stochastic_grad_descent(X, y, objective_func, gradient_func, 
                            alpha=0.01, aug_mag=1,lambda_reg=10**-2, num_epoch=1000,obj_with_reg=False):
    """
    In this question you will implement stochastic gradient descent with regularization term

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        alpha - string or float, step size in gradient descent
                NOTE: In SGD, it's not a good idea to use a fixed step size. Usually it's set to 1/sqrt(t) or 1/t
                if alpha is a float, then the step size in every step is the float.
                if alpha == "1/sqrt(t)", alpha = 1/sqrt(t).
                if alpha == "1/t", alpha = 1/t.
        lambda_reg - the regularization coefficient
        num_epoch - number of epochs to go through the whole training set

    Returns:
        theta_hist - the history of parameter vector, 3D numpy array of size (num_epoch, num_instances, num_features)
                     for instance, theta in epoch 0 should be theta_hist[0], theta in epoch (num_epoch) is theta_hist[-1]
        loss hist - the history of loss function vector, 2D numpy array of size (num_epoch, num_instances)
    """
    num_instances, num_features = X.shape[0], X.shape[1]+1
    theta = np.ones(num_features) #Initialize theta

    theta_hist = np.zeros((num_epoch, num_instances, num_features)) #Initialize theta_hist
    loss_hist = np.zeros((num_epoch, num_instances)) #Initialize loss_hist
    
    if isinstance(alpha, float):
        alphas = np.asarray([alpha for element in np.arange(num_epoch)])
    elif alpha == "1/sqrt(t)":
        alphas = 0.03*(1/np.sqrt(np.arange(1,num_epoch+1)))
    elif alpha == "1/t":
        alphas = 0.03*(1/np.arange(1,num_epoch+1))
    elif alpha == "optional":
        alphas = 0.005/(1+0.005*lambda_reg*np.arange(1,num_epoch+1))

    for epoch in range(num_epoch):
        X,y = shuffle(X,y)
        alpha = alphas[epoch]
    
        for i in range(X.shape[0]):
            Xi = X[i]
            yi = y[i]
            
            if obj_with_reg:
                loss_hist[epoch,i] = objective_func(X, y, theta,aug_mag=aug_mag,lambda_reg=lambda_reg)
            else:
                loss_hist[epoch,i] = objective_func(X, y, theta,aug_mag=aug_mag)
                
            theta_hist[epoch,i] = theta
            
            grad = gradient_func(Xi, yi, theta, lambda_reg=lambda_reg,aug_mag=aug_mag)
            theta = theta - alpha*grad 
        #if epoch%10 == 0:
        #    print('epoch: ',epoch,', loss: ',loss_hist[epoch,-1])
    return theta_hist,loss_hist
    
```

```python
alpha_list = [0.03,0.01,0.005,'1/sqrt(t)','1/t','optional']
fig,axes = plt.subplots(nrows=2,ncols=3)
for (i,alpha) in enumerate(alpha_list):
    theta_hist,loss_hist = stochastic_grad_descent(X_train, y_train,
                                               compute_regularized_square_loss,
                                               compute_regularized_square_loss_gradient, 
                                               alpha=alpha, aug_mag=1,lambda_reg=10**-2, num_epoch=1000,
                                              obj_with_reg=True)
    ax = axes[1*(i>2),i%3]
    ax.plot(np.log10(loss_hist[:,-1]))
    ax.set_xlabel('epochs')
    ax.set_ylabel('log10(L2 loss)');
    ax.set_title('step size = '+str(alpha))
plt.subplots_adjust(wspace=0.5,hspace=0.5)
```

```python
def average_stochastic_grad_descent(X, y, objective_func, gradient_func, 
                            alpha=0.01, aug_mag=1,lambda_reg=10**-2, num_epoch=1000,obj_with_reg=False):
    """
    In this question you will implement stochastic gradient descent with regularization term

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        alpha - string or float, step size in gradient descent
                NOTE: In SGD, it's not a good idea to use a fixed step size. Usually it's set to 1/sqrt(t) or 1/t
                if alpha is a float, then the step size in every step is the float.
                if alpha == "1/sqrt(t)", alpha = 1/sqrt(t).
                if alpha == "1/t", alpha = 1/t.
        lambda_reg - the regularization coefficient
        num_epoch - number of epochs to go through the whole training set

    Returns:
        theta_hist - the history of parameter vector, 3D numpy array of size (num_epoch, num_instances, num_features)
                     for instance, theta in epoch 0 should be theta_hist[0], theta in epoch (num_epoch) is theta_hist[-1]
        loss hist - the history of loss function vector, 2D numpy array of size (num_epoch, num_instances)
    """
    num_instances, num_features = X.shape[0], X.shape[1]+1
    theta = np.ones(num_features) #Initialize theta

    theta_hist = np.zeros((num_epoch, num_instances, num_features)) #Initialize theta_hist
    loss_hist = np.zeros((num_epoch, num_instances)) #Initialize loss_hist
    
    if isinstance(alpha, float):
        alphas = np.asarray([alpha for element in np.arange(num_epoch)])
    elif alpha == "1/sqrt(t)":
        alphas = 0.05*np.append(1/np.sqrt(np.arange(1,num_epoch)),1/np.sqrt(num_epoch))
    elif alpha == "1/t":
        alphas = 0.05*np.append(1/np.arange(1,num_epoch),1/num_epoch)

    for epoch in range(num_epoch):
        X,y = shuffle(X,y)
        alpha = alphas[epoch]
    
        for i in range(X.shape[0]):
            Xi = X[i]
            yi = y[i]
            
            if obj_with_reg:
                loss_hist[epoch,i] = objective_func(X, y, theta,aug_mag=aug_mag,lambda_reg=lambda_reg)
            else:
                loss_hist[epoch,i] = objective_func(X, y, theta,aug_mag=aug_mag)
                
            M = epoch*X.shape[0]+i+1                
            theta = (M-1)/M * theta_hist[epoch,i] + 1/M*theta # moving average

            theta_hist[epoch,i] = theta
            
            grad = gradient_func(Xi, yi, theta, lambda_reg=lambda_reg,aug_mag=aug_mag)
            theta = theta - alpha*grad 

        #if epoch%10 == 0:
        #    print('epoch: ',epoch,', loss: ',loss_hist[epoch,-1])
    return theta_hist,loss_hist
```

```python
theta_hist,loss_hist = average_stochastic_grad_descent(X_train, y_train,
                                               compute_regularized_square_loss,
                                               compute_regularized_square_loss_gradient, 
                                               alpha=0.005, aug_mag=1,lambda_reg=10**-2, num_epoch=1000,
                                          obj_with_reg=True)
```

```python
theta_mag = np.reshape(np.linalg.norm(theta_hist,axis=-1),-1)

fig,ax = plt.subplots(ncols=2)
ax[0].plot(np.log10(loss_hist[:,-1]))
ax[0].set_xlabel('epochs')
ax[0].set_ylabel('log10(L2 loss)');
ax[1].plot(theta_mag)
ax[1].set_xlabel('steps')
ax[1].set_ylabel('log10(theta norm)');
plt.subplots_adjust(wspace=0.5)
```
